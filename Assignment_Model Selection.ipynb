# Pre_processing

```python
#to check the null values

df.isna().sum()
```

```python
#checking for duplicates

df.duplicated().sum()
```

```python
#to fill the null in AGE we use median method &ffill is used in embarkedb

df['Age'] = df['Age'].fillna(df['Age'].median())
df['Embarked'] = df['Embarked'].fillna(method = 'ffill')
```

```python
df.drop('Cabin', axis = 1 , inplace = True)
```

```python
df.isna().sum()
```

```python
df
```

```python
#changing the obj type using one hot encoding

df = pd.get_dummies(df, columns =['Sex','Embarked'])
```

```python
#droping both name and ticket

df.drop(['Name','Ticket'], axis = 1, inplace =True)
```

```python
#selecting the most probable columns fpor outlier detection

num_col = ['Fare','Age','SibSp','Pclass','Parch']
```

```python
for i in num_col:
    plt.figure()
    plt.boxplot(df[i])
    plt.title(i)
```

```python
# here the outliers are Fare ,Age ,SibSp ,Parch
```

```python
#we are leaving both parch &sibsp as it is
```

```python
#fare has large no of outliers so removing them

Q1 = np.percentile(df['Fare'],25,method = 'midpoint')
Q2 = np.percentile(df['Fare'],50,method = 'midpoint')
Q3 = np.percentile(df['Fare'],75,method = 'midpoint')

print("q1 is",Q1)
print("q2 is",Q2)
print("q3 is",Q3)

IQR = Q3-Q1
low_lim = Q1 - 1.5*IQR
up_lim = Q3 + 1.5*IQR

print('IQR is',IQR)
print('low_lim is',low_lim)
print('up_lim is',up_lim)
```

```python
#catching the outliers in fare and removing them
outlier = []
for x in df ["Fare"]:
    if((x>up_lim)or(x<low_lim)):
        outlier.append(x)
```

```python
outlier
```

```python
removed = df["Fare"]>up_lim
df.loc[removed].index
df.drop(df.loc[removed].index,inplace = True)
```

```python
print('skewness value of Age: ',df['Age'].skew())
```

```python
# since the skewness value of age is less than 1 , they have almost to normal distribution
```

# Creating Training Model

```python
x = df.drop('Survived' , axis=1)
y = df['Survived']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=42,test_size=0.2)
```

# KNN

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score,confusion_matrix,recall_score,precision_score,classification_report,f1_score
```

```python
metric_k=[]
neighbors=np.arange(3,20)
for k in neighbors:
    classifier=KNeighborsClassifier(n_neighbors=k,metric='euclidean')
    classifier.fit(x_train,y_train)
    y_pred_knn=classifier.predict(x_test)
    acc=accuracy_score(y_test,y_pred_knn)
    metric_k.append(acc)
```

```python
metric_k
```

```python
#plotting the graph to find the k value

plt.plot(neighbors,metric_k,'o-')
plt.xlabel('k')
plt.ylabel('accuracy')
plt.show()
```

```python
#here lets take k as 17 as it gives the highest accuracy
knn=KNeighborsClassifier(n_neighbors=17)
knn.fit(x_train,y_train)
y_pred_knn=knn.predict(x_test)
```

```python
print('Accuracy= ',accuracy_score(y_test,y_pred_knn))
```

```python
confusion_matrix(y_test,y_pred_knn)
```

# SVM

```python
from sklearn.svm import SVC
svm_cls=SVC(kernel='linear')
svm_cls=svm_cls.fit(x_train,y_train)
y_pred_svm1 = svm_cls.predict(x_test)
```

```python
accuracy_score(y_test,y_pred_svm1)
```

```python
confusion_matrix(y_test,y_pred_svm1)
```
